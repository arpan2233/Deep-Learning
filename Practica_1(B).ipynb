```python
import zipfile
zip_ref=zipfile.ZipFile('iris.zip')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
file_path='Iris.csv'
df=pd.read_csv(file_path)
```

```python
# Dropping 'Id' column for visualization
df_numeric = df.drop(columns=['Id'])
```

```python
print (df_numeric);
```

```python
#  1. Histogram
plt.figure(figsize=(10,6))
df_numeric.hist(bins=20, edgecolor='black', figsize=(10,6))
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()
```

```python
#  2. Stem-and-Leaf Plot
stemgraphic.stem_graphic(df["SepalLengthCm"])
plt.show()
```

```python
#  3. Box Plot
plt.figure(figsize=(10,6))
sns.boxplot(data=df_numeric.drop(columns=['Species']))
plt.title("Box Plot of Iris Features")
plt.show()
```

```python
 # 4. Pie Chart
species_counts = df['Species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors=['red',
'blue', 'green'])
plt.title("Species Distribution in Iris Dataset")
plt.show()
```

```python
#  5. Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species',
palette='Dark2', s=100)
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
#  6. Contour Plot
plt.figure(figsize=(8,6))
sns.kdeplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', fill=True)
plt.title("Contour Plot of Sepal Length vs. Sepal Width")
plt.show()
```
